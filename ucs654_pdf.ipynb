```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("india_airquality.csv", encoding="latin1")
df.head(5)
```

```python

x = df['no2'].dropna().values
```

```python
r =102317253

ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)
```

```python
z = x + ar * np.sin(br * x)

mu = np.mean(z)
variance = np.var(z)
lam = 1 / (2 * variance)
c = np.sqrt(lam / np.pi)
```

```python
print("Learned Parameters:")
print("mu (μ) =", mu)
print("lambda (λ) =", lam)
print("c =", c)
```
